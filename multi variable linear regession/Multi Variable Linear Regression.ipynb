### Theory
Regression: It is the prediction of a dependent variable given some independent variable.

Example: Given features such as location, experience, cost of living, level of education and role predicting the salaries of employees fall under the category of Regression.

It should be noted that before applying any machine learning algorithm, a process called feature selection is done so there exist some form of relationship between the selected features which needs to be found as well as to reduce the number of features on which the model has to be trained on thus reducing complexity and training time.

Linear Regression:
It falls under supervised learning where training data along with the correct predicted values are required to train the model.

In this method we assume that the dependent variable y has a linear relationship with the independent variable x = {x1, x2, …., xn} weighted with w= {w0, w1, w2, …, wn} in form below.

                                        y = w0 + w1*x1 + w2*x2+ ……. + wn*xn

The weights are called as the regression coefficients which are determined by the machine learning algorithm. w0 is considered as the bias which essentially provides an offset.


### Working of Algorithm:

Since multiple variable linear regression is being implemented the relationship is assumed as follows

                                        y = w0 + w1*x1 + w2*x2 + ... + wn*xn

similar to Linear Regrssion with a single variable

## problem statement
BMI calculator : By predicting new values of x which are not in the training set we can calculate the BMI of any person given his age , height and weight so that he/she can take better care of their health and can know if they are prone to
disease such diabeties.

Input: x is the set of features that comprise of Age, Height and Weigth of a person
output : y is the Body mass index of that person



```python
# importing libraries

import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
%matplotlib inline
```

```python
# Assuming Hypothesis: BMI = w0 + w1*age + w2*height + w3*weight

df = pd.read_csv('BMI.csv')
df.columns = ['Sex','Age','Height','Weight',"BMI"]
x = df.drop(["Sex","BMI"],axis = 1)
y = df["BMI"]

# Normalization
x["Age"] = (x["Age"] - np.mean(x["Age"]))/np.std(x["Age"])
x["Height"] = (x["Height"] - np.mean(x["Height"]))/np.std(x["Height"])
x["Weight"] = (x["Weight"] - np.mean(x["Weight"]))/np.std(x["Weight"])

# Splitting into train and test data (20% as test data)
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
model = LinearRegression(fit_intercept=True)

# training the model
model.fit(X_train,y_train)
```

```python
# predicting values using the text data
pred = model.predict(X_test)

# plotting BMI vs Age , BMI vs Height, BMI vs Weight
plt.title("Actual data vs predicted data")
plt.scatter(X_test['Age'],pred)
plt.scatter(X_test['Age'],y_test)
plt.legend(["actual","predicted"])
plt.xlabel("age normalized")
plt.ylabel("BMI")
plt.show()

plt.scatter(X_test['Height'],pred)
plt.scatter(X_test['Height'],y_test)
plt.legend(["actual","predicted"])
plt.xlabel("height normalized")
plt.ylabel("BMI")
plt.show()

plt.scatter(X_test['Weight'],pred)
plt.scatter(X_test['Weight'],y_test)
plt.legend(["actual","predicted"])
plt.xlabel("weight normalized")
plt.ylabel("BMI")
plt.show()


# printing the mean square error
print('Mean squared error: %.2f'%mean_squared_error(y_test, pred))
```

```python
#summary of the weights 
weights = model.coef_
bias = model.intercept_
print('w0 value: ', bias)
print('w1 value: ', weights[0])
print('w2 value: ', weights[1])
print('w3 value: ', weights[2])
```
